```python
def count_unique(football_stat, cols):
    for col in cols:
        print('\n' + 'For column ' + col)
        print(football_stat[col].value_counts())

cat_cols = ['Overall','Value', 'Preferred Foot', 'Wage']
count_unique(football_stat, cat_cols)
```

# Visualizing Players Data for Regression

```python
def plot_bars(football_stat, cols):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) #Define the plot area.
        ax = fig.gca() # define axis
        counts = football_stat[col].value_counts() # find the counts for each unique category
        counts.plot.bar(ax = ax, color = 'red') # Use the plot.bar method on the counts data frame
        ax.set_title('Number of players by' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel('Number of players')# Set text for y axis
        plt.show()
    
plot_cols = ['Position','Work Rate','Preferred Foot']
plot_bars(football_stat, plot_cols)    
```

# Histograms

```python
def plot_bars(football_stat, cols, bins = 10): #"bin" (or "bucket") the range of values—that is, divide the entire range of values into a series of intervals
    for col in cols:
        fig = plt.figure(figsize=(6,6)) #Define the plot area.
        ax = fig.gca() # define axis
        football_stat[col].plot.hist(ax = ax, bins = bins) # Use the plot.hist method on subset of the data frame
        ax.set_title('Histogram of' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel('Number of players')# Set text for y axis
        plt.show()
    
num_cols = ['Jersey Number','Overall','Dribbling']
plot_bars(football_stat, plot_cols) 
```

# Kernel density plots and introduction to Seaborn

```python
def plot_density_hist(football_stat, cols, bins = 10, hist = True):
    for col in cols:
        sns.set_style("whitegrid")
        sns.distplot(football_stat[col], bins = bins, rug=True, hist = hist)
        plt.title('Histogram of ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel('Number of players')# Set text for y axis
        plt.show()
        
plot_density_hist(football_stat, num_cols)
```

# Two dimensional plots

Scatter plots are widely used to examine the relationship between two variables. In this case, the plots created are of some features vs. Potential

```python
def plot_scatter(football_stat, cols, col_y = 'Finishing'):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # define plot area
        ax = fig.gca() # define axis   
        football_stat.plot.scatter(x = col, y = col_y, ax = ax)
        ax.set_title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel(col_y)# Set text for y axis
        plt.show()
        
regression_cols = ['Overall','Potential', 'Jersey Number']
        
plot_scatter(football_stat, regression_cols)        
```

# Deal with overplotting

Use transparency of the points to allow the view to see though points. With mild over plotting this approach can be quite effective.
Countour plots or 2d density plots show the density of points, much as a topographic map shows elevation. Generating the contours has high computational complexity and making this method unsuitable for massive datasets.
Hexbin plots are the two-dimensional analog of a histogram. The density of the shading in the hexagonal cells indicates the density of points. Generating hexbins is computationally efficient and can be applied to massive datasets

The code in the cell below modifies the scatter plot function used previously to add a transparency argument. In statistical graphics, alpha, the inverse of transparency is specified; alpha = 1.0 is opaque, alpha = 0.0 is perfectly transparent. The code in the cell below uses low alpha of 0.2 (high transparency). Execute this code and examine the results.

```python
def plot_scatter(football_stat, cols, col_y = 'Finishing', alpha = '1.0'):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # define plot area
        ax = fig.gca() # define axis   
        football_stat.plot.scatter(x = col, y = col_y, ax = ax, alpha = alpha)
        ax.set_title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel(col_y)# Set text for y axis
        plt.show()
        
regression_cols = ['Overall','Potential','Jersey Number']
        
plot_scatter(football_stat, regression_cols, alpha=0.1)  
```

```python
def plot_desity_2d(football_stat, cols, col_y = 'Finishing', kind ='kde'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.jointplot(col, col_y, data=football_stat, kind=kind)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()

plot_desity_2d(football_stat, regression_cols)    
```

```python
plot_desity_2d(football_stat, regression_cols, kind = 'hex')   
```

# Relation between categorical and numeric variables


Box plots which highlight the quartiles of a distribution. Not surprisingly, the box plot contains a box. The range of the inner two quartiles are contained within the box. The lenght of the box shows the interquartile range. A line within the box shows the median. Wiskers extend for the maximum of 1.5 times the interquartile range or the extreme value of the data. Outliers beyond the wiskers are shown in a symbol.
Violine plots which are a variation on the 1d KDE plot. Two back to back KDE curves are used to show the density estimate.

```python
def plot_box(football_stat, cols, col_y = 'Finishing'):
    for col in cols:
        fig = plt.figure(figsize=(10,10)) # define plot area
        sns.set_style("whitegrid")
        sns.boxplot(col, col_y, data=football_stat)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
cat_cols = ['Position','Work Rate','Preferred Foot']
plot_box(football_stat, cat_cols)    
```

The preferred Foot is not much different.
there are some noticeable differences between the Finishing by category. ie Position and work-rate

```python
def plot_violin(football_stat, cols, col_y = 'Finishing'):
    for col in cols:
        fig = plt.figure(figsize=(10,10)) # define plot area
        sns.set_style("whitegrid")
        sns.violinplot(col, col_y, data=football_stat)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
plot_violin(football_stat, cat_cols) 
```

The interpretation of the violin plots is similar to the box plots. However, a bit more detail on distributions is visible. The area of each violin is the same on each plot display. Notice also that a type of box plot is visible inside each violin plot.

# Use aesthetics to add project additional dimensions

In this section, plot aesthetics are used to project additional dimensions. Some aesthetics are useful only for categorical variables, while others are useful for numeric variables. Keep in mind that not all plot aesthetics are equally effective. Tests of human perceptions have shown that people are very good as noticing small differences in position. This fact explains why scatter plots are so effective. In rough order of effectiveness these aesthetics are:

Marker shape is an effective indicator variable category. It is critical to select shapes which are easily distinguished by the viewer.
Marker size shows values of a numeric variable. Be careful, as size is the span across the marker, not the area.
Marker color is useful as an indicator of variable category. Color is the least effective of these three aesthetics in terms of human perception. Colors should be chosen to appear distinct. Additionally, keep in mind that many people, particularly men are red-green color blind.

# Marker shape

```python
def plot_scatter_shape(football_stat, cols, shape_col = 'Preferred Foot', col_y = 'Finishing', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    unique_cats = football_stat[shape_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            temp = football_stat[football_stat[shape_col] == cat]
            sns.regplot(col, col_y, data=temp, marker = shapes[i], label = cat,
                        scatter_kws={"alpha":alpha}, fit_reg = False, color = 'blue')
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()
            
num_cols = ['Overall']
plot_scatter_shape(football_stat, num_cols) 
```

# Marker size

```python
def plot_scatter_size(football_stat, cols, shape_col = 'Preferred Foot', size_col = 'Potential',
                            size_mul = 0.000025, col_y = 'Finishing', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    unique_cats = football_stat[shape_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            temp = football_stat[football_stat[shape_col] == cat]
            sns.regplot(col, col_y, data=temp, marker = shapes[i], label = cat,
                        scatter_kws={"alpha":alpha, "s":size_mul*football_stat[size_col]**2}, 
                        fit_reg = False, color = 'blue')
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()

num_cols = ['Overall']
plot_scatter_shape(football_stat, num_cols)
```

### Color

```python
def plot_scatter_shape_size_col(football_stat, cols, shape_col = 'Preferred Foot', size_col = 'Potential',
                            size_mul = 0.000025, color_col = 'Work Rate', col_y = 'Finishing', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    colors = ['green', 'blue', 'orange', 'magenta', 'gray'] # specify distinctive colors
    unique_cats = football_stat[shape_col].unique()
    unique_colors = football_stat[color_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            for j, color in enumerate(unique_colors):
                temp = football_stat[(football_stat[shape_col] == cat) & (football_stat[color_col] == color)]
                sns.regplot(col, col_y, data=temp, marker = shapes[i],
                            scatter_kws={"alpha":alpha, "s":size_mul*temp[size_col]**2}, 
                            label = (cat + ' and ' + color), fit_reg = False, color = colors[j])
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()

num_cols = ['Overall']
plot_scatter_shape(football_stat, num_cols)
```
